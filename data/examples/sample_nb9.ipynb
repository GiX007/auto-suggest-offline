```python
import pandas as pd
tips = pd.read_csv("tips.csv")
```

```python
# Drop missing values
cleaned = tips.dropna()
```

```python
# Melt tips by gender and time
melted = pd.melt(cleaned, id_vars=['sex', 'time'], value_vars=['total_bill', 'tip'])
melted.head()
```

```python
# Group by time and variable
grouped = melted.groupby(['time', 'variable'])['value'].mean()
grouped.head()
```
