```python
import pandas as pd
flights = pd.read_csv("flights.csv")
flights.head()
```

```python
# Fill missing values
flights_filled = flights.fillna(0)
```

```python
# Drop rows missing departure delay
flights_cleaned = flights_filled.dropna(subset=['passengers'])
```

```python
# Calculate how many passengers raveled each year on average
grouped = flights_cleaned.groupby('year')['passengers'].mean()
print(grouped)
```
