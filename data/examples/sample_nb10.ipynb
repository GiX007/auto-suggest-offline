```python
import pandas as pd
penguins = pd.read_csv("penguins.csv")
```

```python
species_lookup = pd.DataFrame({
    'species': ['Adelie', 'Chinstrap', 'Gentoo'],
    'type': ['Type A', 'Type B', 'Type C']
})
species_lookup.head()
```

```python
# Fill missing values
filled = penguins.fillna(method='ffill')
filled.head()
```

```python
# Merge penguins with species type info
merged = pd.merge(filled, species_lookup, on='species', how='left')
merged.head()
```

```python
# Group by species type and calculate average body mass
grouped = merged.groupby('type')['body_mass_g'].mean()
grouped.head()
```
